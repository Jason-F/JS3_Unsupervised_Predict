# Filtering Methods

## Content Based Filtering

#### What is Content Based Filtering

The Content based filtering method is based on the descriptive data available for a particular item or product. This method will look at items that are similar in their description and context establish recommendations based on that. The users prior preferences are taken into consideration to find products that match their historic preferences. If for example, a user likes to watch Toy Story, then they will be recommended movies that are very similar to it, like the rest of the Toy Story series and other Pixar movies.

#### Why Content Based Filtering is not used

Content based filtering has some fundamental flaws due to it being based on one users information and not multiple, which will over specify recommendations. Another flaw is due to it being inherently retrospective which will not help the user discover new content outside their existing preferences. An example of this issue is if a user watches drama movies, the content based filtering method will then recommend them drama movies but this will limit what kind of content the user will be exposed to. Perhaps they would enjoy comedy or horror too? This not only makes for a suboptimal user experience but also from a business standpoint, it will reduce the content watched and exposure to products which will negatively affect the companies wanting their content on this businesses recommender system.

```python
# Create a copy of a dataframe
movies = df_movies.copy()
```

Now that we have made a copy of the data frame to work with, we will then merge two dataframes, to better implement our method

```python
df_1 = df_imdb[['movieId','title_cast','director', 'plot_keywords']]
df_1 = df_1.merge(movies[['movieId', 'genres', 'title']], on='movieId', how='inner')
df_1.head()
```

Now that we have successfully merged the dataframes, we will need to make the data more usable by:
-	Converting data types to strings
-	Remove space between names
-	Remove the lines between the title_cast and get only the primary names, which in this case will be the first three
-	Remove the lines between the plot_keywords  and get only the primary keywords which in this case will be the first 5
-	Remove the lines between genres


```python
# Change data to strings for string handling
df_1['title_cast'] = df_1.title_cast.astype(str)
df_1['plot_keywords'] = df_1.plot_keywords.astype(str)
df_1['genres'] = df_1.genres.astype(str)
df_1['director'] = df_1.director.astype(str)

# Removing lines between names
df_1['director'] = df_1['director'].apply(lambda x: "".join(x.lower() for x in x.split()))
df_1['title_cast'] = df_1['title_cast'].apply(lambda x: "".join(x.lower() for x in x.split()))

# Removing lines between the actors' full names and getting only the first three names
df_1['title_cast'] = df_1['title_cast'].map(lambda x: x.split('|'))

# Removing lines between the plot keywords' and getting only the first five words
df_1['plot_keywords'] = df_1['plot_keywords'].map(lambda x: x.split('|'))
df_1['plot_keywords'] = df_1['plot_keywords'].apply(lambda x: " ".join(x))

# Removing lines between the genres 
df_1['genres'] = df_1['genres'].map(lambda x: x.lower().split('|'))
df_1['genres'] = df_1['genres'].apply(lambda x: " ".join(x))

df_1.head()
```

Well need to further modify the dataset before we can finalise our content based filter.

We will need to use cosine similarity which is a technique used for measuring the similarity between vectors. This takes two vectors and calculates the cosine of the angle between them.

```python
#Converting the title_cast column from list to string
df_1['liststring'] = df_1['title_cast'].apply(lambda x: ','.join(map(str, x)))

#Removing the commas between the string in each row
df_1['liststring'] = df_1['liststring'].replace(',',' ', regex=True)

#Use keywords, cast(liststring), diector and genres column to use as our features
df_features = df_1[['liststring','director','plot_keywords','genres']]

#Combine the features columns into  single string
df_1['combined_features'] = df_features['liststring'] +' '+ df_features['director'] +' '+ df_features['plot_keywords'] +' '+ df_features['genres']

#Use CountVectorizer() object for getting the cv matrix.
cv =CountVectorizer()
cv_matrix = cv.fit_transform(df_1['combined_features'])

#Getting cosine similarity matrix from the cv matrix
sim_score = cosine_similarity(cv_matrix,cv_matrix)

df_1.set_index('title', inplace = True)
indices = pd.Series(df_1.index)
```

Now we can display our score using the cosine similiarty matrix

```python
print(sim_score)
```

In order to implement our content based filter we will now create a method that will take in a recommendation via an inputted movie title and return n based recommendations (where n is specified via an int variable inputted after the movie title). 

```python
# This method will make n recommendations based off the movie title input
def content_filter(title,n,sim_score = sim_score):
    movies_recommend = []
    
    # Index of the movie that matches the title
    idx = indices[indices == title].index[0]

    # Series with the similarity scores in descending order
    score_series = pd.Series(sim_score[idx]).sort_values(ascending = False)

    # getting the indexes of the 10 most similar movies
    top_n_indexes = list(score_series.iloc[1:n+1].index)
    
    # populating the list with the titles of the best n matching movies
    for i in top_n_indexes:
        movies_recommend.append(list(df_1.index)[i])
        
    return movies_recommend
```

```python
content_filter('Toy Story (1995)',10)
```

We can see the user has been returned 10 movie titles that are very similar to the users input. This matches what we described earlier that content based filters inherently recommend items that are very similar and do not recommend items that deviate much.
As a result, we can conclude that a content based filter method is not usable for our movie recommendation system. We will now go on to implement a Collaborative Filtering method.


## Collaborative Filtering

#### What is Collaborative Filtering

The Collaborative Filtering technique filters recommendations for a user based on the reactions by other similar users, resulting in recommendations gathered and generated by many users instead of just one.

This method will use a large user base and look for subsets of them with similar preferences to a particular user. It will base the recommendations on the similarities in their preferences and provide a wider range of content to the user.

#### Why Collaborative Filtering is used

It is a better method to use for our recommender system since to it can conduct feature learning on its own, which will enable it to learn which features are best to use and it will only grow more useful in time. Collaborative filtering users other user ratings to create decisions and adapts to the users interests which will inevitably change over time. Collaborative Filtering is vastly superior to the rigid and restrictive content based filtering method.

# Model Selection

<b>Singular Value Decomposition(SVD)</b>  is a collaborative filtering method, using a factorization of a real or complex matrix that generalizes the eigendecomposition, which only exists for square normal matrices, to any m\times n matrix via an extension of the polar decomposition.



SVD is a matrix factorisation technique, which reduces the number of features in a dataset. Using SVD as a collaborative filtering technique, it utilizes a matrix structure where each row represents a user, and each column represents an item. The elements of this matrix are the ratings that are given to items by users.

## Using SVD with hyperparameter tuning.

We will now conduct parameter tuning of our SVD model to enhance the recommendation results.

The parameters we will be tuning:
- <b>n_factors:</b> The number of factors. Default is 100
	- we will change to 300
- <b>lr_all:</b>  The learning rate for all parameters. Default is 0.005.
	- we will change to 0.006
- <b>reg_all:</b> The regularization term for all parameters. Default is 0.02.
	- we will change to 0.04
- <b>n_epochs:</b> The number of iteration of the SGD procedure. Default is 20.
	- we will change to 60
- <b>nit_std_dev:</b> The standard deviation of the normal distribution for factor vectors initialization. Default is 0.1
	- we will change to 0.04
